```python
import pandas as pd


def clean_data(file_path='art_auction_valuation/data.txt'):
    """ 
    Essential data cleaning: 
    - Reads the dataset
    - Removes unwanted columns
    - Cleans country names
    - Converts numerical columns
    - Standardizes material names
    - Removes missing values 
    """
    # Read dataset (TSV format)
    df = pd.read_csv(file_path, sep='\t', encoding='utf-8')

    # Drop unnecessary column (if it exists)
    if 'Unnamed: 0' in df.columns:
        df.drop(columns=['Unnamed: 0'], inplace=True)

    # Remove invalid country names (must be alphabetic)
    df = df[df['country'].str.isalpha() == True]

    # Convert height & width to numeric (remove non-numeric values)
    df['height'] = pd.to_numeric(df['height'], errors='coerce')
    df['width'] = pd.to_numeric(df['width'], errors='coerce')

    # Remove rows where price, height, or width is missing
    df.dropna(subset=['price', 'height', 'width'], inplace=True)

    # Standardize material names (lowercase)
    df['material'] = df['material'].str.lower()

    return df


# Run the function
df_cleaned = clean_data()
df_cleaned  # Check the cleaned data
```

```python
import requests
from bs4 import BeautifulSoup

# Sotheby's Auction Results URL
base_url = "https://www.sothebys.com/en/results"
headers = {'User-Agent': 'Mozilla/5.0'}

response = requests.get(base_url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

# Extract auction links with proper formatting
auction_links = []
for link in soup.find_all('a', href=True):
    href = link['href']

    # Check if the link is a relative URL or already a full URL
    if href.startswith("http"):
        auction_links.append(href)
    elif href.startswith("/"):
        auction_links.append("https://www.sothebys.com" + href)

# Remove duplicates & print
auction_links = list(set(auction_links))
print(auction_links[:5])  # Preview first 5 auction links
```

```python
# Function to fix closeYear column
def fix_year(year):
    if pd.isna(year):  # Handle NaN values
        return None
    year = int(year)  # Convert to integer
    if year > 2000:  # Already in four-digit format
        return year
    elif year >= 50:  # Assume 1950-1999
        return 1900 + year
    else:  # Assume 2000-2049
        return 2000 + year


# Apply function to transform closeYear properly
df_christies["sale_year"] = df_christies["closeYear"].apply(fix_year)

# Drop unnecessary columns
df_christies.drop(columns=["closeYear"], inplace=True)

# Save cleaned dataset
df_christies.to_csv("christies_cleaned.csv", index=False)

# Check result
print(df_christies[["sale_year"]].head())
```

```python


# Standardize column names
df_sothebys.rename(columns={
    'Title': 'title', 'Artist': 'artist', 'Currency': 'currency',
    'Selling Price': 'price', 'low_estimate': 'low_estimate',
    'high_estimate': 'high_estimate', 'Sold': 'sold',
    'Auction Name': 'auction_title', 'Date': 'sale_date'
}, inplace=True)

df_christies.rename(columns={
    'Art Title': 'title', 'Artist': 'artist', 'Currency': 'currency',
    'price': 'price', 'low_estimate': 'low_estimate',
    'high_estimate': 'high_estimate', 'Sold': 'sold',
    'Auction Title': 'auction_title', 'close date': 'sale_date'
}, inplace=True)

# Convert sale date to datetime format
df_sothebys['sale_date'] = pd.to_datetime(
    df_sothebys['sale_date'], errors='coerce')
df_christies['sale_date'] = pd.to_datetime(
    df_christies['sale_date'], errors='coerce')

# Extract year of creation from "DoC" in Christie's dataset
df_christies['year_created'] = df_christies['DoC'].apply(
    lambda x: int(str(x)[:4]) if pd.notnull(x) and str(x).isdigit() else None)

# Extract height & width from Christie's dataset dimensions
if 'Dimentions' in df_christies.columns:
    df_christies[['height', 'width']] = df_christies['Dimentions'].str.extract(
        r'(\d+\.?\d*)\s*x\s*(\d+\.?\d*)').astype(float)

# Drop unnecessary columns
df_sothebys.drop(columns=['Unnamed: 0', 'Signed',
                 'Category', 'Image'], inplace=True, errors='ignore')
df_christies.drop(columns=['DoC', 'Dimentions',
                  'Alive?'], inplace=True, errors='ignore')

# Merge datasets
df_merged = pd.concat([df_sothebys, df_christies], ignore_index=True)

# Save the merged dataset
df_merged.to_csv("merged_auction_data.csv", index=False)

# Display summary of final dataset
print(df_merged.info())
print(df_merged.head())
```
